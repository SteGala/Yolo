```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("data.csv")

data = data[data.Time > "2024-07-09 21:47:37"]
averages = []

for start in range(0, len(data), 324):
    end = start + 300
    chunk = data.iloc[start:end]
    chunk = chunk[chunk["yolo-server-798f5c5578-ll75h / default / PKG"] != 0]
    chunk = chunk.drop(columns=["Time"])
    avg = chunk.mean()
    averages.append(avg)

# Convert the list of averages to a DataFrame
avg_df = pd.DataFrame(averages)

# Plot the stacked bar plot
avg_df.plot(kind='bar', stacked=True)
plt.xlabel('Chunk Index')
plt.ylabel('Average Value')
plt.title('Stacked Bar Plot of Averages for Each Chunk')
plt.show()
    
```
